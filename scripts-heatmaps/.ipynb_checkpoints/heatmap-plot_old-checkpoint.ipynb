```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
INPUT_FOLDER = '../data/wrangled/merged/'
OUTPUT_FOLDER = '../figs/'
DATE = '2016-07-25'
FORMATS = ['svg','pdf']
```

```python
# color values
POSITIVE = '#BA053C'# dark pink
NEGATIVE = '#FFE9B2' #yellow
STERILE = '#C2EDEF' #light blue
FILLER =  '#FF9C09' #orange

COLOR_VALS = np.array([POSITIVE, NEGATIVE, STERILE, FILLER])
```

```python
# cats for pivot table
INDEXES = ['lesion_tissue_type','lesion_locus','lesion_id','lesion','z','lesion_dissem_type']
COLUMNS = ['barcode_dissem_type','barcode_id','barcode','max_lesion_z']
VALUES = ['present']
```

```python
def find_files():
    files = np.array([])
    for directory, sub, fnames in os.walk(INPUT_FOLDER):
        for f in fnames:
            if f.split("-")[0]=='merged' and f[-4:]=='.csv': 
                fpath = "%s%s"%(directory, f)
                files = np.append(files, [fpath])
    return files
files = find_files()
```

```python
def construct_barcode_refs(group):
    keys = ['num_lesions','gran_pos','max_lesion_z', 'lesions_list','lesions_z']
    num_lesions = len(group)
    grans = group.loc[group.lesion_tissue_type=='Gran']
    gran_pos = True if len(grans) > 0 else False
    max_lesion_z = max(grans.z.values) if len(grans)>0 else max(group.z.values)
    lesions_list = group.lesion.values
    lesions_z = group.z.values
    values = [num_lesions, gran_pos, max_lesion_z, lesions_list, lesions_z]
    return pd.Series(values,keys)

barcodes = testd.groupby('barcode').apply(construct_barcode_refs)
barcodes.sort_values( by=['gran_pos','max_lesion_z','num_lesions'], ascending=False, inplace=True )
barcodes['barcode_id'] = range(len(barcodes))
for prop in ['num_lesions','gran_pos','max_lesion_z', 'barcode_id']:
    testd.loc[:,prop] = testd.barcode.apply(lambda x: barcodes.loc[x,prop])
```

```python
def get_data(csv):
#     cols_needed = [ 'lesion_tissue_type','lesion_locus','lesion_id','lesion','barcode','dissem_type','barcode_id','present'] 
    full_data = pd.read_csv(csv).drop('Unnamed: 0',axis=1)
    data = full_data
    #     data = full_data[cols_needed]
    return data
```

```python
def create_pivot_table(data):
    # create pivot table
    pivotdf = pd.pivot_table(data, index=INDEXES, columns=COLUMNS, values=VALUES, fill_value=1)
    # sort columns and index before taking out unnecessary ones
    pivotdf.columns.index = pivotdf.columns.sortlevel(level=['dissem_type','barcode_id'])
    pivotdf.sort_index(level=['lesion_tissue_type','lesion_id'],inplace=True)
    pivotdf.columns = pivotdf.columns.droplevel([0,'barcode_id'])
    pivotdf.reset_index(level=['lesion_id'], inplace=True, drop=True)

    # change values in table to reflect lost and sterile lesions
    if 'lost' in pivotdf.columns: pivotdf.loc[pivotdf.lost.x==3, ['contained','disseminated']] = 3
    if 'sterile' in pivotdf.columns: pivotdf.loc[pivotdf.sterile.x==2, ['contained','disseminated']] = 2
    pivotdf = pivotdf[['contained','disseminated']]

    return pivotdf
```

```python
def amend_pivot_table_index(pivotdf):

    # get good labels
    labels_dissem_type, labels_barcode = [p.values() for p in pivotdf.columns.labels]
    label_idx = np.where((labels_dissem_type < 2))
    good_labels = [labels_dissem_type[label_idx], labels_barcode[label_idx] ]
    
    # get good levels
    level_dissem_types, level_barcodes = pivotdf.columns.levels
    level_dissem_types = level_dissem_types[np.where([(level_dissem_types!='lost') & (level_dissem_types!='sterile')])[1]]
    level_barcodes = level_barcodes[level_barcodes!='x']
    good_levels = [level_dissem_types, level_barcodes]
    
    # get names (for sake of consistency, added 'good')
    good_names = pivotdf.columns.names
    
    #create and apply new MultiIndex
    new_pivot_col_index = pd.MultiIndex(levels=good_levels, labels=good_labels, names=good_names)
    pivotdf.columns = new_pivot_col_index
    return pivotdf
```

```python
def make_heatmap_arr_data(key, pivotdf):
    
    tissue, locus, dissem = key
    # mask replace order is important here
    arr = pivotdf.loc[(tissue,locus),dissem].copy()
    mask = arr.values==3
    arr.replace(to_replace=3, value=2, inplace=True)

    # construct color map
    temp_out = np.unique(arr.values)
    present_keys = temp_out[temp_out!=3]
    cvals = [FILLER if not(i in present_keys) else COLOR_VALS[i] for i in range(max(present_keys)+1)]
    palette = sns.blend_palette(cvals, n_colors=len(cvals), as_cmap=True)
    
    # construct tick labels
    yticklabels = np.array(["%s"%(l) for l in arr.index.values])
    
    ytl_rev_idx = np.array(sorted(range(len(yticklabels)), reverse=True))
    yticklabels = yticklabels[ytl_rev_idx]
    return arr, mask, palette, yticklabels
```

```python
def plot_heatmap(key, pivotdf, monkey):

    arr, mask, palette, yticklabels = make_heatmap_arr_data(key, pivotdf)
    tissue, locus, dissem = key
    
    # actually plot the thing
    f, ax = plt.subplots(1)
    ax = sns.heatmap( arr, vmin=0, mask=mask, cmap=palette, ax=ax, square=True, linewidths=1, cbar=False )
    
    # axes formatting and such
    ax.set_xlabel("")
    ax.set_ylabel("")
    xticklabels = [t.get_text() for t in ax.get_xticklabels()]
    
    ax.tick_params(axis='both', which='major', pad=8)
    ax.set_xticklabels(xticklabels, rotation=90, ha='center', fontsize='x-large', weight='semibold')
    ax.set_yticklabels(yticklabels, rotation=0, ha='right', fontsize='x-large', weight='semibold')
    
    title = " ".join([monkey, tissue, locus if locus!='na' else "", dissem])
#     ax.set_title(title, size='large', y=1.02)
    w = len(xticklabels)*.6
    h = len(yticklabels)
    f.set_size_inches(w, h)
    # YAY!
    return f, ax, title
```

```python
def iterate_pivoted(pivotdf, monkey):
    graphs = {}

    tiss_loc_keys = np.unique(pivotdf.reset_index('lesion').index.values)
    dissem_keys = np.unique(pivotdf.columns.get_level_values('dissem_type'))
    for tiss, loc in tiss_loc_keys:
        for dissem in dissem_keys:
            key = (tiss, loc, dissem)
            f, ax, title = plot_heatmap(key, pivotdf, monkey)
            graphs[key] =  f,ax
    return graphs
```

```python
all_graphs = {}
pivoted_dfs = {}

if not os.path.isdir("../data/pivoted-noculturesstuffs"): os.mkdir("../data/pivoted-noculturesstuffs")
for directory, sub, files in os.walk(INPUT_FOLDER):
    for f in files:
        path = "%s%s"%(directory,f)
        monkey = f[:-4].split("-")[1]
        data = get_data(path)
        if monkey in exclude: 
            data = data.loc[data.lesion.isin(exclude[monkey])==False]
        pivoted = create_pivot_table(data)
        pivoted = amend_pivot_table_index(pivoted)
        pivoted.to_csv("../data/pivoted-noculturesstuffs/pivoted-noculturesstuffs-%s.csv"%monkey)
        pivoted_dfs[monkey] = pivoted
        graphs = iterate_pivoted(pivoted, monkey)
        all_graphs[monkey] = graphs
```

```python
# for noticklabels
OTHER_OUTPUT_FOLDER = "../figs/figs_20160725-noticklabels-no-culturestuffs/"
if not os.path.isdir(OTHER_OUTPUT_FOLDER): os.mkdir(OTHER_OUTPUT_FOLDER)

for monkey, graphs in all_graphs.items():
    monkey_path = "%s/%s/"%(OTHER_OUTPUT_FOLDER, str(monkey))
    if not os.path.isdir(monkey_path): 
        os.mkdir(monkey_path)
    for fmt in FORMATS:
        fmt_path = monkey_path+fmt+"/"
        if not os.path.isdir(fmt_path):
            os.mkdir(fmt_path)
    for key, (f,ax) in graphs.items():
        tissue_type, locus, dissem = key
        if tissue_type != 'LN': ax.set_xticklabels([])
        if dissem == 'contained': ax.set_yticklabels([])
        title = locus if tissue_type == 'Gran' else tissue_type
        for fmt in FORMATS:
            file_path = "%s%s/%s-%s-%s.%s"%(monkey_path, fmt, monkey, title, dissem, fmt)
            f.savefig(file_path)
```
